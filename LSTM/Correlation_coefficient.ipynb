```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

table = pd.read_csv('table_output.csv',encoding='utf-8')

corr_coe = table.corr()
plt.figure(figsize=(8,8))
plt.title("CORRELATION OF FEATURES", fontsize=15)
sns.heatmap(corr_coe,cmap="Greens")
plt.xticks(rotation=(-60), fontsize=10)
plt.yticks(fontsize=10)
plt.show()
print(corr_coe)
```

```python
table_selected=table.drop(["bitcoin_transactions", "bitcoin_sent", "mine_a_block_min", "Block_size", 
                  "Weighted_gold_price", "generatedCoin", "bitcoin_sent_inNum","Volume_litecoin"],axis=1)
```

```python
corr_coe = table_selected.corr()
plt.figure(figsize=(8,8))
plt.title("CORRELATION OF FEATURES", fontsize=15)
sns.heatmap(corr_coe,cmap="Greens")
plt.xticks(rotation=(-60), fontsize=10)
plt.yticks(fontsize=10)
plt.show()
print(corr_coe)
```
