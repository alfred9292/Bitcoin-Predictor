```python
from pandas import read_csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = read_csv('Regulation.csv',header=0)
date = []
for i in range(0,data.shape[0]):
    x = data.iloc[i].values[0]
    y = x.split('/')[0]+'/'+x.split('/')[1]
    date.append(y)
```

```python
result=Counter(date)
date1=[]
fre=[]    
dict = {}
for key in date:
    dict[key] = dict.get(key, 0) + 1
print (dict)

for i in dict:
    date1.append(i)
    fre.append(dict[i])
print(date1)
print(fre)
```

```python
plt.figure(figsize=(20,5))
plt.ylabel('Frequency')
plt.title(' Regulation News Frequency Based On Month In a Year')
plt.xticks(rotation=90)
plt.ylim([0,15])
for x,y in enumerate(fre):
    plt.text(x,y+0.1,'%s' %round(y,1),ha='center')
plt.bar(date1, fre)
plt.show()
```

```python
data = read_csv('Regulation.csv',header=0)
us=[]
eu=[]
cn=[]
jp=[]
for i in range(0,data.shape[0]):
    x = data.iloc[i].values[0]
    y = data.iloc[i].values[2]
    z = x.split('/')[0]+'/'+x.split('/')[1]
    if'U.S.'in y or 'SEC'in y :
     us.append(z)
    if'EU'in y or 'European'in y:
     eu.append(z)
    if'Chinese'in y or 'China'in y:
     cn.append(z)
    if'Japan'in y:
     jp.append(z)
        
```

```python
date1=[]
fre1=[]    
dict1 = {}
for key in us:
    dict1[key] = dict1.get(key, 0) + 1
#print (dict)

for i in dict1:
    date1.append(i)
    fre1.append(dict1[i])
print(date1)
print(fre1)
plt.figure(figsize=(20,5))
plt.ylabel('Frequency')
plt.title('U.S. Regulation Frequency')
plt.xticks(rotation=90)
for x,y in enumerate(fre1):
    plt.text(x,y+0.1,'%s' %round(y,1),ha='center')
plt.bar(date1, fre1)
plt.show()
```
