```python
from pandas import read_csv
```

```python
#calculate sentiment score for regulation news and combine the results to the data output.
data = read_csv('data output4.csv',header=0)
data1 = read_csv('Regulation result.csv',header=0)
date={}

for i in range(0,data1.shape[0]):
    d=data1.iloc[i].values[0]
    s=data1.iloc[i].values[1]
    x1=s.replace('[','')
    x2=x1.replace(']','')
    date[d]=(x2)
for i in range(0,data.shape[0]):
    x = data.iloc[i].values[0]
    if x in date:
        data.loc[i,'Regulation Sentiment']=date[x]
        
    else:
        data.loc[i,'Regulation Sentiment']=data.loc[i-1,'Regulation Sentiment']
    
    print(data.loc[i,'Regulation Sentiment'])
    
data.to_csv('data output4.csv')

```
