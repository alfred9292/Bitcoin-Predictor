```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
import re
#extract data from Coinmarketcap
url = "https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20150617&end=20190415"
content = requests.get(url).content
soup = BeautifulSoup(content,'html.parser')
table = soup.find('table', {'class': 'table'})

data = [[td.text.strip() for td in tr.findChildren('td')] 
        for tr in table.findChildren('tr')]

df = pd.DataFrame(data)
df.drop(df.index[0], inplace=True) # first row is empty
df[0] =  pd.to_datetime(df[0]) # date
for i in range(1,7):
    df[i] = pd.to_numeric(df[i].str.replace(",","").str.replace("-","")) # some vol is missing and has -
df.columns = ['Date','Open','High','Low','Close','Volume','Market Cap']
df.set_index('Date',inplace=True)
df.sort_index(inplace=True)
df.to_csv('moving_average.csv')
```

```python
#extract bitcoin data from Coinmarketcap
url = "https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20150701&end=20190316"
content = requests.get(url).content
soup = BeautifulSoup(content,'html.parser')
table = soup.find('table', {'class': 'table'})

data = [[td.text.strip() for td in tr.findChildren('td')] 
        for tr in table.findChildren('tr')]

df_bit = pd.DataFrame(data)
df_bit.drop(df_bit.index[0], inplace=True) # first row is empty
df_bit[0] =  pd.to_datetime(df_bit[0]) # date
for i in range(1,7):
    df_bit[i] = pd.to_numeric(df_bit[i].str.replace(",","").str.replace("-","")) # some vol is missing and has -
df_bit.columns = ['Date','Open','High','Low','Close','Volume','Market Cap']
df_bit.set_index('Date',inplace=True)
df_bit.sort_index(inplace=True)
```

```python
#extract data from bitinfoCharts - number of bitcoin transaction per day
url = 'https://bitinfocharts.com/comparison/price-btc-sma14.html'
response = requests.get(url)
soup = BeautifulSoup(response.text,'lxml')
script_tag = soup.findAll('script')[5]
script_text = script_tag.text
pattern = re.compile(r'\[new Date\("\d{4}/\d{2}/\d{2}"\),\d*\w*\]')
records = pattern.findall(script_text)
print(records)

def parse_record(record):
    date = record[11:21]
    value = record[24:-1]
    return [date,value]

transactions = []

for record in records:
    transactions.append(parse_record(record))
    
bitcoin_transactions = pd.DataFrame(transactions)
bitcoin_transactions = bitcoin_transactions.set_index(pd.DatetimeIndex(bitcoin_transactions[0]))
bitcoin_transactions= bitcoin_transactions.drop([0],axis=1)
bitcoin_transactions.columns = ['RSI']
bitcoin_transactions.to_csv('moving_average.csv')
#bitcoin_transactions.head()
#bitcoin_transactions.tail()
```

```python
#extract data from bitinfoCharts - number of bitcoin active address
url = 'https://bitinfocharts.com/comparison/price-btc-rsi14.html'
response = requests.get(url)
soup = BeautifulSoup(response.text,'lxml')
script_tag = soup.findAll('script')[5]
script_text = script_tag.text
pattern = re.compile(r'\[new Date\("\d{4}/\d{2}/\d{2}"\),\d*\w*\]')
records = pattern.findall(script_text)
print(records)

def parse_record(record):
    date = record[11:21]
    value = record[24:-1]
    return [date,value]

Num_bitcoin_address = []

for record in records:
    Num_bitcoin_address.append(parse_record(record))
    
Num_bitcoin_address = pd.DataFrame(Num_bitcoin_address)
Num_bitcoin_address = Num_bitcoin_address.set_index(pd.DatetimeIndex(Num_bitcoin_address[0]))
Num_bitcoin_address= Num_bitcoin_address.drop([0],axis=1)
Num_bitcoin_address.columns = ['rsi']
Num_bitcoin_address.tail()
#Num_bitcoin_address.to_csv('rsi.csv')
```

```python
#extract data from bitinfoCharts - number of tweets per day
url = 'https://bitinfocharts.com/comparison/bitcoin-tweets.html'
response = requests.get(url)
soup = BeautifulSoup(response.text,'lxml')
script_tag = soup.findAll('script')[5]
script_text = script_tag.text
pattern = re.compile(r'\[new Date\("\d{4}/\d{2}/\d{2}"\),\d*\w*\]')
records = pattern.findall(script_text)

def parse_record(record):
    date = record[11:21]
    value = record[24:-1]
    return [date,value]

Num_tweets = []

for record in records:
    Num_tweets.append(parse_record(record))
    
Num_tweets = pd.DataFrame(Num_tweets)
Num_tweets = Num_tweets.set_index(pd.DatetimeIndex(Num_tweets[0]))
Num_tweets= Num_tweets.drop([0],axis=1)
Num_tweets.columns = ['Num_tweets']
#Num_tweets.tail()
```

```python
#extract data from bitinfoCharts - number of tweets per day
url = 'https://bitinfocharts.com/comparison/bitcoin-difficulty.html'
response = requests.get(url)
soup = BeautifulSoup(response.text,'lxml')
script_tag = soup.findAll('script')[5]
script_text = script_tag.text
pattern = re.compile(r'\[new Date\("\d{4}/\d{2}/\d{2}"\),\d*\w*\]')
records = pattern.findall(script_text)

def parse_record(record):
    date = record[11:21]
    value = record[24:-1]
    return [date,value]

bitcoin_difficulty = []

for record in records:
    bitcoin_difficulty.append(parse_record(record))
    
bitcoin_difficulty = pd.DataFrame(bitcoin_difficulty)
bitcoin_difficulty = bitcoin_difficulty.set_index(pd.DatetimeIndex(bitcoin_difficulty[0]))
bitcoin_difficulty= bitcoin_difficulty.drop([0],axis=1)
bitcoin_difficulty.columns = ['mining_difficulty']
#bitcoin_difficulty.tail()
```

```python
#read data from AppleStore.csv into dataframe format
data_bitcoin = df_bit
data_litecoin = df
data_gold = pd.read_csv('gold.csv',encoding='utf-8')
data_oil = pd.read_csv('oil.csv',encoding='utf-8')
data_usd_euro = pd.read_csv('USD_EURO.csv',encoding='utf-8')
data_wiki_view = pd.read_csv('pageviews.csv',encoding='utf-8')
#extract from https://data.bitcoinity.org/bitcoin/difficulty/5y?t=l
#show dimensionality
print(data_bitcoin.shape)
print(data_litecoin.shape)
print(data_gold.shape)
print(data_usd_euro.shape)
print(data_oil.shape)
print(data_wiki_view.shape)
#change to time-Series
#data_bitcoin = data_bitcoin.set_index(pd.DatetimeIndex(data_bitcoin['Date']))
data_gold = data_gold.set_index(pd.DatetimeIndex(data_gold['DATE']))
data_oil = data_oil.set_index(pd.DatetimeIndex(data_oil['date']))
data_usd_euro = data_usd_euro.set_index(pd.DatetimeIndex(data_usd_euro['date']))
data_wiki_view = data_wiki_view.set_index(pd.DatetimeIndex(data_wiki_view['Date']))

#pre-processing bitcoin data
data_bitcoin['Weighted_bitcoin_price'] = round(data_bitcoin.loc[:,["Open", "Close", "High", "Low"]].mean(axis=1),2)
data_bitcoin = data_bitcoin.drop(["Open", "Close", "High", "Low", "Market Cap"], axis=1)
columnsTitles=["Weighted_bitcoin_price","Volume"]
data_bitcoin=data_bitcoin.reindex(columns=columnsTitles)
#print(data_bitcoin.head())

#pre-processing litecoin data
data_litecoin['Weighted_litecoin_price'] = round(data_litecoin.loc[:,["Open", "Close", "High", "Low"]].mean(axis=1),2)
data_litecoin = data_litecoin.drop(["Open", "Close", "High", "Low", "Market Cap", "Volume"], axis=1)
#print(data_litecoin.head())

#pre-processing gold data
data_gold['Weighted_gold_price'] = round(data_gold.loc[:,["OPEN", "CLOSING", "HIGH", "LOW"]].mean(axis=1),2)
data_gold = data_gold.drop(["DATE", "OPEN", "CLOSING", "HIGH", "LOW"], axis=1)
#print(data_gold.head())

#pre-processing oil data
data_oil=data_oil.drop(["date"],axis=1)
data_oil.columns = ['Oil_price']
#print(data_oil.head())

#pre-processing usd/euro data
data_usd_euro=data_usd_euro.drop(["date"],axis=1)
data_usd_euro.columns = ['usd/euro']
#print(data_usd_euro.head())

#pre-processing number of wikipedia views
data_wiki_view=data_wiki_view.drop(["Date"],axis=1)
data_wiki_view.columns = ['Num_Wiki_view']
#print(data_wiki_view.head())

if np.any(data_gold.isnull()) != True:
    print("There is no empty cell in the dataset")
if np.any(data_gold[data_gold.index.duplicated()])!= True:
    print("There is no duplicate information in the dataset")
```
