```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
import re
#extract data from Coinmarketcap
url = "https://coinmarketcap.com/currencies/litecoin/historical-data/?start=20140302&end=20190401"
content = requests.get(url).content
soup = BeautifulSoup(content,'html.parser')
table = soup.find('table', {'class': 'table'})

data = [[td.text.strip() for td in tr.findChildren('td')] 
        for tr in table.findChildren('tr')]

df = pd.DataFrame(data)
df.drop(df.index[0], inplace=True) # first row is empty
df[0] =  pd.to_datetime(df[0]) # date
for i in range(1,7):
    df[i] = pd.to_numeric(df[i].str.replace(",","").str.replace("-","")) # some vol is missing and has -
df.columns = ['Date','Open','High','Low','Close','Volume','Market Cap']
df.set_index('Date',inplace=True)
df.sort_index(inplace=True)
```

```python
#extract bitcoin data from Coinmarketcap
url = "https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20150701&end=20190316"
content = requests.get(url).content
soup = BeautifulSoup(content,'html.parser')
table = soup.find('table', {'class': 'table'})

data = [[td.text.strip() for td in tr.findChildren('td')] 
        for tr in table.findChildren('tr')]

df_bit = pd.DataFrame(data)
df_bit.drop(df_bit.index[0], inplace=True) # first row is empty
df_bit[0] =  pd.to_datetime(df_bit[0]) # date
for i in range(1,7):
    df_bit[i] = pd.to_numeric(df_bit[i].str.replace(",","").str.replace("-","")) # some vol is missing and has -
df_bit.columns = ['Date','Open','High','Low','Close','Volume','Market Cap']
df_bit.set_index('Date',inplace=True)
df_bit.sort_index(inplace=True)
```

```python
#extract data from bitinfoCharts - number of bitcoin transaction per day
url = 'https://bitinfocharts.com/comparison/bitcoin-transactions.html'
response = requests.get(url)
soup = BeautifulSoup(response.text,'lxml')
script_tag = soup.findAll('script')[5]
script_text = script_tag.text
pattern = re.compile(r'\[new Date\("\d{4}/\d{2}/\d{2}"\),\d*\w*\]')
records = pattern.findall(script_text)

def parse_record(record):
    date = record[11:21]
    value = record[24:-1]
    return [date,value]

transactions = []

for record in records:
    transactions.append(parse_record(record))
    
bitcoin_transactions = pd.DataFrame(transactions)
bitcoin_transactions = bitcoin_transactions.set_index(pd.DatetimeIndex(bitcoin_transactions[0]))
bitcoin_transactions= bitcoin_transactions.drop([0],axis=1)
bitcoin_transactions.columns = ['bitcoin_transactions']
#bitcoin_transactions.head()
#bitcoin_transactions.tail()
```
