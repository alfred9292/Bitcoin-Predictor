```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.ensemble import RandomForestRegressor
import pandas
from pandas import Series 
from pandas import read_csv
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn import preprocessing 
```

```python
df = read_csv('data output.csv')
array=df.values
x=array[:,2:15]
y=array[:,1]
clf=RandomForestRegressor(n_estimators=1000, random_state=0, n_jobs=-1)
clf.fit(x,y)
```

```python
# random forest regressor as selector
sfs = SFS(clf, 
          k_features=5, 
          forward=True, 
          floating=False, 
          scoring='neg_mean_squared_error',
          cv=5)

sfs = sfs.fit(x, y)
```

```python
fig = plot_sfs(sfs.get_metric_dict(), kind='std_err')

plt.title('Sequential Forward Selection (w. StdErr)')
plt.grid()
plt.show()
```

```python
result=list(sfs.k_feature_idx_)
print(result)
```

```python
# x ranges from the 3rd column to the 22nd column
df = read_csv('Final.csv')
array=df.values
x=array[:,2:21]
y=array[:,1]
x=preprocessing.scale(x)
#print(x)
#print(y)
clf=RandomForestRegressor(n_estimators=1000, random_state=0, n_jobs=-1)
clf.fit(x,y)
```

```python
sfs = SFS(clf, 
          k_features=5, 
          forward=True, 
          floating=False, 
          scoring='neg_mean_squared_error',
          cv=5)

sfs = sfs.fit(x, y)
```

```python
fig = plot_sfs(sfs.get_metric_dict(), kind='std_err')

plt.title('Sequential Forward Selection (w. StdErr)')
plt.grid()
plt.show()
```

```python
'''
feature selection for the top 5 features 
these features are: mining difficulty, litecoin price, number of wiki views, number of tweets and 3 days moving average.

''' 
result=list(sfs.k_feature_idx_)
print(result)
```

```python
'''
linear regression as selector,feature selection for the top 5 features 
these features are: mine a block min, number of wili views, 3 days moving average,14 days moving average and privacy ans security news sentiment.
'''
clf=LinearRegression()
sfs = SFS(clf, 
          k_features=5, 
          forward=True, 
          floating=False, 
          scoring='neg_mean_squared_error',
          cv=10)

sfs = sfs.fit(x, y)
result=list(sfs.k_feature_idx_)
print(result)
fig = plot_sfs(sfs.get_metric_dict(), kind='std_err')

plt.title('Sequential Forward Selection (w. StdErr)')
plt.grid()
plt.show()
```
