```python
import pandas as pd 
import numpy as np 
import csv
data = read_csv('WSJ_news.csv',header=0)
outcome=[]
dic={}
d={}
list1=[]
list2=[]
list3=[]
for i in range(0,data.shape[0]):
    x = str(data.iloc[i].values[1])
    y = data.iloc[i].values[5]
    if x in dic:
        dic[x].append(float(y))
    else:
        dic[x]=[float(y)]

for key in dic:
    result=round(np.mean(dic[key]),2)
    d[key]=[result]
    outcome.append(result)
#print(d)

for key in d.keys():
    list1.append(key)
for value in d.values():
    list2.append(value)
for i in range(0,len(list2)):
    x1=str(list2[i]).replace('[','')
    x2=x1.replace(']','')
    list3.append(x2)

dataframe = pd.DataFrame({'Date':list1,'Average Sentiment':list3 })
dataframe.sort_values("Date",inplace=True)
dataframe.to_csv('wsj daily sentiment.csv')
```
