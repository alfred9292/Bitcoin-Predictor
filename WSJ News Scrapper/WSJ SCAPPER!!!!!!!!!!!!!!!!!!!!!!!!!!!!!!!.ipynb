```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import calendar
```

```python
result=[]
result1=[]
result2=[]
result3=[]
result5=[]
final=[]
num=''

# in total there are 23 pages. in the loop,just update the last digit to refresh the page so that the scrapper will go through all the search results
for k in range(1,23):
    print(k)
    if k==1:
        url = "https://www.wsj.com/search/term.html?KEYWORDS=bitcoin&min-date=2015/07/01&max-date=2019/04/28&isAdvanced=true&daysback=4y&meta=bitcoin&andor=AND&sort=date-desc&source=wsjarticle,sitesearch,press,newswire,wsjpro"
   
        wbdata = requests.get(url).text
        soup = BeautifulSoup(wbdata,'lxml')
        news_titles = soup.select("h3.headline") #get news title

        for n in news_titles:
            title = n.get_text()
            result.append(title)
            #print(title)

        category = soup.select("div.category") # get news category
        for n in category:
            cate = n.get_text()
            result1.append(cate)
            #print(cate)

        date = soup.select("time.date-stamp-container") #get news publication date in YYYY/DD/MM format 
        for n in date:
            time = n.get_text()
            result2.append(time)
            #print(result2)
        for i in range(0,len(result2)):
            day=result2[i].split(' ')[1]
            finalday=day.replace(',', '')
            year=result2[i].split(' ')[2]
            month=result2[i].split(' ')[0]
            finalmonth=month.replace('.', '')
            #print(finalmonth)
            if finalmonth in list(calendar.month_name):
                num=list(calendar.month_name).index(finalmonth)
            else:
                num=list(calendar.month_abbr).index(finalmonth)
            print(num)
            news_date=str(year)+'/'+str(num)+'/'+str(finalday)
            result5.append(news_date)
            print(result5)
            print(news_date)
            #print(finalmonth)
            #print(finalday)
            #print(year)


        summary = soup.select("div.summary-container") # get news summary
        for n in summary:
            summary = n.get_text()
            result3.append(summary)
            #print(summary)

        date=pd.DataFrame(data=list(result5),columns=['Date']) #format the output csv file
        date['Title'] = result
        date['Summary']=result3
        date['Category']=result1

    
        date.to_csv('WSJ_news.csv') #output the result
        result=[]
        result1=[]
        result2=[]
        result3=[]
        result5=[]
        final=[]
    else:
       
        url= "https://www.wsj.com/search/term.html?KEYWORDS=bitcoin&min-date=2015/07/01&max-date=2019/04/28&isAdvanced=true&daysback=4y&meta=bitcoin&andor=AND&sort=date-desc&source=wsjarticle,sitesearch,press,newswire,wsjpro&page="+str(k)
        print(url)
        wbdata = requests.get(url).text
        soup = BeautifulSoup(wbdata,'lxml')
        news_titles = soup.select("h3.headline")

        for n in news_titles:
            title = n.get_text()
            result.append(title)
            print(title)

        category = soup.select("div.category")
        for n in category:
            cate = n.get_text()
            result1.append(cate)
            #print(cate)

        date = soup.select("time.date-stamp-container")
        for n in date:
            time = n.get_text()
            result2.append(time)
        print(len(result2))
        for i in range(0,len(result2)):
            day=result2[i].split(' ')[1]
            finalday=day.replace(',', '')
            year=result2[i].split(' ')[2]
            month=result2[i].split(' ')[0]
            #print(month)
            finalmonth=month.replace('.', '')
            #print(finalmonth)
            if finalmonth in list(calendar.month_name):
                num=list(calendar.month_name).index(finalmonth)
            else:
                num=list(calendar.month_abbr).index(finalmonth)
            #print(num)
            news_date=str(year)+'/'+str(num)+'/'+str(finalday)
            print(news_date)
            result5.append(news_date)
            #print(news_date)
            #print(finalmonth)
        print(result5)

        summary = soup.select("div.summary-container")
        for n in summary:
            summary = n.get_text()
            result3.append(summary)
        print(len(result3))
        if k<22 and len(result3)<20:
            gap= 20-len(result3)
            print(gap)
            for i in range(20-gap,20):
                result3[i]=result3.append('n/a') # find news that contains no summary
                
        print(len(result3))
    #print(result3)
        

        date=pd.DataFrame(data=list(result5),columns=['Date']) #format the output csv file
        date['Title'] = result
        date['Summary']=result3
        date['Category']=result1

        date.to_csv('WSJ_news.csv',mode='a',header=False) #output the result
        result=[]
        result1=[]
        result2=[]
        result3=[]
        result5=[]
        final=[]
    
```
